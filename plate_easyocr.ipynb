```python
def plate_num_one_by_one(index):
    fig, ax = plt.subplots(1,2, figsize=(12,8))
    img_plate= cv2.imread(img_orig[index])
    reader= easyocr.Reader(['en'])
    text= reader.readtext(img_orig[index])
    print(text)
    for i in range(len(text)):
        coord= np.array(text[i][0])
        if cv2.contourArea(coord)> 10000 :
            img_rect= cv2.rectangle(img_plate.copy(),  text[i][0][0], text[i][0][2], (234,98,55), 3)
            img_text= cv2.putText(img_rect, text[i][1], (50,50), cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,255), 6)
            ax[0].imshow(img_plate)
            ax[1].imshow(img_text)
```
